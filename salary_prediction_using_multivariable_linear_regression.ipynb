```python
from word2number import w2n
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df=pd.read_csv("hiring.csv")
```

```python
df
```

```python
df.experience=df.experience.fillna("zero")
```

```python
df
```

```python
df.experience = df.experience.apply(w2n.word_to_num)
```

```python
df
```

```python
me=df['test_score(out of 10)'].mean()
```

```python
me
```

```python
import math
mea=math.floor(me)
```

```python
mea
```

```python
df['test_score(out of 10)']=df['test_score(out of 10)'].fillna(mea)
```

```python
df
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(df[['experience','test_score(out of 10)','interview_score(out of 10)']],df['salary($)'])
```

```python
reg.coef_
```

```python
reg.intercept_
```

# Predict the salary for the employee whose experience is 2 years and test_score(out of 10) is 9 and interview_score(out of 10) is 6

```python
reg.predict([[2,9,6]])
```
