```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("homeprices1.csv")
```

```python
df
```

```python
df.head()
```

# In the bedrooms column there is NaN so we have to manage it.We will fill that with Median of bedroom columns.

```python
import math
bedroom=math.floor(df['bedrooms'].median())
print(bedroom)
```

```python
df.bedrooms.fillna(bedroom)
df.bedrooms=df.bedrooms.fillna(bedroom)
df
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(df[['area','bedrooms','age']],df['price'])
```

```python
reg.coef_
```

```python
reg.intercept_
```

# Now predict the Price of (1)-3000 sq ft area,3 bedrooms,40 years old.  (2)-2500 sq ft area,4 bedrooms,5 years old


```python
reg.predict([[3000,3,40]])
```
